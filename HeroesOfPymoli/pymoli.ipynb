```python
import pandas as pd
import numpy as np
purchase_data = "Resources/purchase_data.csv"
pymoli_df = pd.read_csv(purchase_data)
pymoli_df.head()
```

```python
player_count=len(pymoli_df["SN"].unique())
player_count
summary_table = pd.DataFrame([{"Total_Players": player_count}])
summary_table
```

```python
unique_items=len(pymoli_df["Item ID"].unique())
unique_items


# Avg Price
avg_price=pymoli_df["Price"].mean()
avg_price

#No of Purchases
total_purchase=pymoli_df["Purchase ID"].count()
total_purchase

#
Total_revenue=pymoli_df["Price"].sum()
Total_revenue
```

```python
summary_data=pd.DataFrame([{"Number of Unique Items":unique_items,
                          "Avg Price":round(avg_price,2),
                          "No of Purchases":total_purchase,
                          "Total_revenue":Total_revenue}])
summary_data
```

```python
count_gender =pymoli_df["Gender"].value_counts()
count_gender
```

```python
gender_table = pymoli_df[["SN", "Gender"]]
gender_unique = gender_table.drop_duplicates()
gender_unique.head()
count_gender =gender_unique["Gender"].value_counts()
count_gender

mgsumm=pd.DataFrame(count_gender)
mgsumm["Percentage by total gender"]=mgsumm["Gender"]/player_count*100
mgsumm



    




```

```python
grouped_gender_df = pymoli_df.groupby(['Gender'])
grouped_gender_df.count().head(10)
```

```python
purchase_count=grouped_gender_df['Purchase ID'].count()
purchase_count
gender_data_summary=pd.DataFrame(purchase_count)

gender_data_summary

gender_data_summary["avg_purchase_price"]=grouped_gender_df['Price'].mean()
gender_data_summary["total_purchase_price"]=grouped_gender_df["Price"].sum()
gender_data_summary["avg_total_purchase_per_person"]=grouped_gender_df["Price"].sum()/count_gender
gender_data_summary
```

```python

age_bins = [0, 9,14, 19, 24, 29,34,39,44]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]




```

```python
pymoli_df["Age Group"] = pd.cut(pymoli_df["Age"], bins=age_bins, labels=group_labels)
pymoli_df.head()



ageGroup = pymoli_df.groupby('Age Group').nunique()
ageGroup['Percentage of Players'] = ((ageGroup['SN']/player_count)*100).map('{:.2f}'.format)
ageGroup.rename(columns={"SN": "Total Count"},inplace=True)
ageGroup.filter(items=['Total Count', 'Percentage of Players'])


```

```python
age_table = pymoli_df[["SN", "Age Group"]]
age_table.head()
age_unique = gender_table.drop_duplicates()

pymoli_df['SN Total'] = pymoli_df.groupby(['SN'])['Price'].transform(sum)
pymoli_df.head()
tol_per_age = pymoli_df.drop_duplicates(subset='SN', keep='first').groupby('Age Group').sum()['SN Total']

tol_per_person = pymoli_df.drop_duplicates(subset='SN', keep='first').groupby('Age Group').mean()['SN Total']
tol_per_person
#count_age= age_table.groupby('Age Group').nunique()
#count_age

```

```python

purchase_count = pymoli_df.groupby('Age Group').count()['Purchase ID']
summary_price = pymoli_df.groupby('Age Group').mean()['Price']
Total_purchase_price=pymoli_df.groupby("Age Group").sum()['Price']


age_summary = pd.DataFrame({'Purchase Count':purchase_count, 'Avg Purchase Price': summary_price,
                            'Total Purchase Value':tol_per_age,'Purchase Total Per Person':tol_per_person})
age_summary['Avg Purchase Price'] = age_summary['Avg Purchase Price'].map('${:.2f}'.format)
age_summary['Total Purchase Value'] = age_summary['Total Purchase Value'].map('${:,.2f}'.format)
age_summary['Purchase Total Per Person'] = age_summary['Purchase Total Per Person'].map('${:.2f}'.format)
age_summary



```

```python
spender_count=pymoli_df.groupby('SN').count()['Purchase ID']
spender_avg=pymoli_df.groupby('SN').mean()[('Price')]
spender_price=pymoli_df.groupby('SN').sum()['Price']

SN_summary = pd.DataFrame({'Purchase Count':spender_count, 'Avg Purchase Price':spender_avg, 
                           'Total Purchase Value':spender_price})

SN_summary.sort_values('Total Purchase Value', ascending=False, inplace=True)
SN_summary.head()
```

```python

item_group = pymoli_df.groupby(['Item ID', 'Item Name'])
purchase_count = item_group['Purchase ID'].count()
item_price = item_group['Price'].mean()
total_purchase_value = item_group['Price'].sum()
item_summary = pd.DataFrame({'Purchase Count':purchase_count, 'Item Price':item_price,'Total Purchase Value':total_purchase_value})
item_summary.sort_values('Purchase Count',ascending=False, inplace=True)

item_summary.head()


```
